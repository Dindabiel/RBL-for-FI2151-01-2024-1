```python
import pandas as pd

df = pd.read_csv("gun_deaths.csv")
print(df)
```

# Visualisasi Data

## 1. Jumlah Kematian Akibat Senjata Api Setiap Tahun

```python
import seaborn as sns
import matplotlib.pyplot as plt

deaths2012 = len(df.loc[df["year"] == 2012])
deaths2013 = len(df.loc[df["year"] == 2013])
deaths2014 = len(df.loc[df["year"] == 2014])

deathbyyear_x = ["2012", "2013", "2014"]
deathbyyear_y = [deaths2012, deaths2013, deaths2014]

plt.plot(deathbyyear_x, deathbyyear_y)
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kematian')
plt.title("Jumlah Kematian Akibat Senjata Api")
plt.show()
```

### Dapat dilihat bahwa ada peningkatan jumlah kematian dari tahun 2012 ke 2013 lalu penurunan dari tahun 2013 ke 2014.

## 2. Komposisi Penyebab Kematian

```python
print(df["intent"].value_counts().plot(kind="pie", ylabel="", title = "Komposisi Penyebab Kematian"))
```

### Dapat dilihat bahwa bunuh diri merupakan penyebab terbesar dari kematian akibat senjata api, diikuti dengan aksi pembunuhan. Sebagian kecil dari kematian akibat senjata api disebabkan karena kecelakaan.

## 3. Jumlah Campur Tangan Polisi pada Kematian

```python
police = len(df.loc[df["police"] == 1])
nopolice = len(df.loc[df["police"] == 0 ])

plot_x = ["Ada", "Tidak"]
plot_y = [police, nopolice]

plt.bar(plot_x, plot_y)
plt.title("Jumlah Campur Tangan Polisi pada Kematian")
plt.ylabel("Jumlah Kematian")
plt.show()
```

### Dapat dilihat bahwa sebagian besar dari kematian akibat senjata api di Amerika Serikat tidak melibatkan pihak kepolisian.

## 4. Komposisi Jenis Kelamin Korban

```python
print(df["sex"].value_counts().plot(kind="pie", ylabel="",title = "Komposisi Jenis Kelamin Korban"))
```

### Dapat dilihat bahwa sebagian besar korban kematian akibat senjata api di Amerika Serikat adalah pria.

## 5. Persebaran Usia Korban

```python
print(df["age"].plot(kind="hist", bins = 100, title = "Persebaran Usia Korban", ylabel = "Jumlah Kematian"))
```

### Dapat dilihat bahwa jumlah kematian akibat senjata api banyak didominasi oleh anak muda berumur sekitar 20 tahun-an

## 6. 10 Usia dengan Jumlah Kematian Tertinggi

```python
ages = df["age"].value_counts()
top10 = ages.iloc[0:10]

plot = top10.plot(kind="barh", title = "10 Usia dengan Jumlah Kematian Tertinggi")

plot.invert_yaxis()

plt.xlabel("Jumlah Kematian")
plt.ylabel("Usia (tahun)")

print(plot)
```

### Dapat dilihat bahwa usia dengan jumlah kematian tertinggi adalah 22 tahun. Dapat dilihat juga bahwa 10 usia dengan jumlah kematian tertinggi hanya berusia sekitar 20 tahun-an

## 7. Komposisi Ras Korban

```python
print(df["race"].value_counts().plot(kind="pie", ylabel="", title="Komposisi Ras Korban"))
```

### Dapat dilihat bahwa sebagian besar korban akibat senjata api terbesar di Amerika Serikat merupakan ras kulit putih.

## 8. Jumlah Penyebab Kematian Setiap Ras

```python
raceintent = df[["intent", "race"]].value_counts()
print(raceintent.plot(kind="bar", ylabel = "Jumlah Kematian", xlabel ="Penyebab Kematian dan Ras Korban", title = "Jumlah Setiap Penyebab Kematian Setiap Ras"))
```

### Dapat dilihat bahwa kasus kematian akibat senjata api terbanyak Amerika Serikat adalah kasus bunuh diri oleh masyarakat ras kulit putih. Kasus pada ras kulit hitam dan hispanik didominasi oleh kasus pembunuhan sedangkan pada ras asia dan pasifik adalah kasus bunuh diri.

## 9. Persebaran Lokasi Kematian Korban

```python
print(df["place"].value_counts().plot(kind="bar", ylabel ="Jumlah Kematian", title="Persebaran Lokasi Kematian Korban"))
```

### Dapat dilihat bahwa jumlah kematian terbesar terjadi di rumah.

## 10. Persebaran Pendidikan Tertinggi Korban

```python
print(df["education"].value_counts().plot(kind="bar", title = "Persebaran Pendidikan Tertinggi Korban", ylabel = "Jumlah Kematian"))
```

### Dapat dilihat bahwa sebagian besar korban memiliki tingkat pendidikan tertinggi HS/GED.

## 11. Angka Kematian Akibat Kecelakaan Tahun 2012-2014

```python
jumlahkecelakaan2012 = len(df.loc[(df["year"] == 2012) & (df["intent"] == "Accidental")])
jumlahkecelakaan2013 = len(df.loc[(df["year"] == 2013) & (df["intent"] == "Accidental")])
jumlahkecelakaan2014 = len(df.loc[(df["year"] == 2014) & (df["intent"] == "Accidental")])

xaxis = ["2012", "2013", "2014"]
yaxis = [jumlahkecelakaan2012, jumlahkecelakaan2013, jumlahkecelakaan2014]

plt.plot(xaxis, yaxis)

plt.xlabel("Tahun")
plt.ylabel("Jumlah Kematian Akibat Kecelakaan")
plt.title("Angka Kematian Akibat Kecelakaan 2012 - 2014")
plt.show()
```

### Dapat dilihat bahwa ada penurunan jumlah kasus kematian akibat senjata api pada tahun 2012 - 2013 lalu ada peningkatan pada tahun selanjutnya.

## 12. Angka Kematian Akibat Senjata Api Pada Usia 15 Hingga 25 Tahun

```python
death15252012 = len(df.loc[(df["age"] >= 15) & (df["age"] <= 25) & (df["year"] == 2012)])
death15252013 = len(df.loc[(df["age"] >= 15) & (df["age"] <= 25) & (df["year"] == 2013)])
death15252014 = len(df.loc[(df["age"] >= 15) & (df["age"] <= 25) & (df["year"] == 2014)])

xaxis = ["2012", "2013", "2014"]
yaxis = [death15252012, death15252013, death15252014]

plt.plot(xaxis, yaxis)

plt.xlabel("Tahun")
plt.ylabel("Jumlah Kematian")
plt.title("Angka Kematian Usia 15 - 25")

plt.show()
```

### Dapat dilihat terdapat penurunan angka kematian usia 15 - 25 tahun pada 2012 - 2013 lalu ada sedikit peningkatan pada 2013 - 2014.

## 13. Jumlah Korban per Jenis Kelamin (Gender) Setiap Ras

```python
df = {'Ras': ["White", "Black", "Hispanic", "Asian/Pacific Islander", "Native American/Native Alaskan"],

    'Pria': [
        len(df.loc[(df["sex"] == "M") & (df["race"] == "White")]),
        len(df.loc[(df["sex"] == "M") & (df["race"] == "Black")]),
        len(df.loc[(df["sex"] == "M") & (df["race"] == "Hispanic")]),
        len(df.loc[(df["sex"] == "M") & (df["race"] == "Asian/Pacific Islander")]),
        len(df.loc[(df["sex"] == "M") & (df["race"] == "Native American/Native Alaskan")])
    ], 
    'Wanita': [
        len(df.loc[(df["sex"] == "F") & (df["race"] == "White")]),
        len(df.loc[(df["sex"] == "F") & (df["race"] == "Black")]),
        len(df.loc[(df["sex"] == "F") & (df["race"] == "Hispanic")]),
        len(df.loc[(df["sex"] == "F") & (df["race"] == "Asian/Pacific Islander")]),
        len(df.loc[(df["sex"] == "F") & (df["race"] == "Native American/Native Alaskan")])
    ]
    }

df = pd.DataFrame(data=df)
df.plot(kind="bar", x = "Ras", y = ["Pria", "Wanita"], stacked = True)
plt.ylabel("Jumlah Kematian")
plt.title("Jumlah Korban per Jenis Kelamin (Gender) Setiap Ras")
plt.show()
```

### Dapat dilihat bahwa untuk setiap ras, korban senjata api selalu didominasi oleh pria.
